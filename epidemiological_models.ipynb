```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
import networkx as nx
import seaborn as sns
import math
plt.rcParams['text.usetex'] = True
```

```python
T_R = 5
mu = 1/T_R
alph = 0.5
c = 0.5
beta = alph * c
N = 10000
S = [N - (1*(1/100)*N)]
I = [1*(1/100)*N]
R = [0]

for t in range(10000):
    P_si = beta*I[t]/N 
    S_out = np.random.binomial(S[t], P_si)
    P_ir = mu
#     print(t, S[t], I[t], R[t])
    S.append( S[t] - S_out)
    I.append( I[t] + S_out - P_ir*I[t])
    R.append( R[t] + P_ir*I[t])
    

plt.figure(figsize = (6, 6))
plt.plot(list(range(100)), S[1:101], marker = '.', label = 'S')
plt.plot(list(range(100)), I[1:101], marker = '.', label = 'I')
plt.plot(list(range(100)), R[1:101], marker = '.', label = 'R')
plt.legend()
plt.title("SIR model: mu = 0.2 beta = 0.4 I0 = 1000 N  = 10000")
```

```python
T_R = 5
mu = 1/T_R
alph = 0.001
lambda_ = 0.99
N = 10000
I = [9990]
S = [10]
R = [0]

for t in range(10000):
    P_is = lambda_*S[t]/N 
    P_sr = alph*(S[t] + R[t])/N
    I_out = np.random.binomial(I[t], P_is)
    S_out = np.random.binomial(S[t], P_sr)
    P_ir = mu
#     print(t, S[t], I[t], R[t])
    I.append( I[t] - I_out)
    S.append( S[t] + I_out - S_out)
    R.append( R[t] + S_out)
    
    
```

```python
plt.figure(figsize = (16, 6))
plt.plot(list(range(10000)), S[0:10000], marker = '.', label = 'S')
plt.plot(list(range(10000)), I[0:10000], marker = '.', label = 'I')
plt.plot(list(range(10000)), R[0:10000], marker = '.', label = 'R')
plt.legend()
plt.title("SIR model: mu = 0.2 beta = 0.4 I0 = 1000 N  = 10000")
```
